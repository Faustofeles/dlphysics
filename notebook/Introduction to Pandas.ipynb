## Dataframe Creation

### Dataframes using Comma-Separated Value (CSV) files

```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/concat_1.csv') # try to see how these are displayed in Spyder IDE
df2 = pd.read_csv('data/concat_2.csv')
df3 = pd.read_csv('data/concat_3.csv')
```

```python
print("Shape: ", df1.shape)
df1 
```

```python
# concatenate rows
# (stack on top of each other)
row_concat = pd.concat([df1, df2, df3])
print("New Shape: ", row_concat.shape)
row_concat
```

```python
# get back column A
print(row_concat['A'])
print("\nData Type of row_concat: ", type(row_concat))
print("Data Type of extracted column A: ", type(row_concat['A']))
```

```python
# concatenate columns
# stack to the left/right
col_concat = pd.concat([df1, df2, df3], axis=1) # axis=0 is default and means row wise operation
print("New Shape: ", col_concat.shape)
col_concat
```

```python
# note when you have the same columns and you concatenate them
# the column name is retained
# so when you subset by that column name, you get them all back
col_concat['A']
```

```python
# rename the column names and concatenate 
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']

concatenated_by_col = pd.concat([df1, df2, df3], axis=1)
concatenated_by_col
```

```python
# what happens if relabeling the columns and concatenating by rows instead?
```

```python
# rename the column names and concatenate 
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']

concatenated_by_row = pd.concat([df1, df2, df3], sort=True)
concatenated_by_row
```

```python
# saving Dataframe into CSV file
concatenated_by_col.to_csv('concatenated.csv')
!ls # this is the bash ls command, not python
```

### Dataframes using Python Dictionaries

```python
# Creating Dataframes using dictionaries
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])

frames = [df1, df2, df3]
result_concat = pd.concat(frames)
result_concat
```

```python
result_append = df1.append(df2)
result_append
```

Here is a good reference for the different methods available in Pandas to merge Dataframes: https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
