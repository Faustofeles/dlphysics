# Introduction to Matplotlib

[**Matplotlib**](https://matplotlib.org/) is a 2D ploting library for Python which produces publication quality figures in a variety of harcopy formats and interactive environments across multiple platforms. It is very easy to use and great for performing visualizations on the go! 

## Basic Set Up 

The first thing to do is to import matplotlib, the convention is as follows: 
```python
import matplotlib.pyplot as plt
```

However, if we want to have our plots appear in our Jupyter Notebook "inline" we need to use the following IPython built-in magic function:

```python
%matplotlib inline 
```

IPython magic functions create an environment in which you can choose something other than the expected result. In our case, the above magic function for Matplotlib creates an environment where our plots appear embedded in our notebook. Here is a fast and easy way to make a plot:

````python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-10, 10, 100)
Y = np.sin(X)

plt.plot(X,Y)
plt.title("Sinusoidal Wave")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")


'''
If we did not have the magic function shown above, 
we would have had to use the function plt.show() 
to display our graph.
'''
````

## Plot Anatomy

The above example was a simple way of making a quick graph. If we want to start creating more complex plots, we need to first study the anatomy of a matplotlib plot. 

![image.png](img/matplotlib_anatomy.png)


````python
x = np.linspace(0, 5, 10)
y = x ** 2
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_xlabel('y')
axes.set_title('title')
````

````python
# You can also create the fig and axes using th subplots function

fig, axes = plt.subplots()
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_xlabel('y')
axes.set_title('title')
````

````python
# this is the reason you want to use subplots() to create your figures
fig, axes = plt.subplots(nrows=1, ncols=2)
# axes[0].set_figsize((12,3))
axes[0].plot(x, y, 'r')
axes[1].plot(x, y, 'g')
````

````python
x = np.linspace(-10, 10, 100)
y = np.power(x, 3)
y1 = 3*x**2
y2 = 6*x

fig, axes = plt.subplots(nrows=1, ncols=3)
axes[0].plot(x, y, 'r')
axes[1].plot(x, y1, 'g')
axes[2].plot(x, y2, 'b')
````
