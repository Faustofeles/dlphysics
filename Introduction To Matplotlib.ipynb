# Introduction to Matplotlib

[**Matplotlib**](https://matplotlib.org/) is a 2D ploting library for Python which produces publication quality figures in a variety of harcopy formats and interactive environments across multiple platforms. It is very easy to use and great for performing visualizations on the go! 

## Basic Set Up 

The first thing to do is to import matplotlib, the convention is as follows: 
```python
import matplotlib.pyplot as plt
```

However, if we want to have our plots appear in our Jupyter Notebook "inline" we need to use the following IPython built-in magic function:

```python
%matplotlib inline 
```

IPython magic functions create an environment in which you can choose something other than the expected result. In our case, the above magic function for Matplotlib creates an environment where our plots appear embedded in our notebook. Here is a fast and easy way to make a plot:

````python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-10, 10, 100)
Y = np.sin(X)

plt.plot(X,Y)
plt.title("Sinusoidal Wave")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")

'''
If we did not have the magic function shown above, 
we would have had to use the function plt.show() 
to display our graph.
'''
````

## Plot Anatomy

The above example was a simple way of making a quick graph. If we want to start creating more complex plots, we need to first study the anatomy of a matplotlib plot. 

![image.png](img/matplotlib_anatomy.png)


````python
x = np.linspace(0, 5, 20)
y = x ** 2
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
````

````python
# You can also create the fig and axes using th subplots function

fig, axes = plt.subplots()
# fig, axes = plt.subplots(figsize=(12,3)) # you can also change the aspect ratio
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
````

## Multiple Plots in a Figure 

````python
# this is the reason you want to use subplots() to create your figures
fig, axes = plt.subplots(nrows=1, ncols=2)
# axes[0].set_figsize((12,3))
axes[0].plot(x, y, 'r')
axes[1].plot(x, y, 'g') 
````

````python
x = np.linspace(-10, 10, 100)
y = np.power(x, 3)
y1 = 3*x**2
y2 = 6*x

# figsize argument sets the aspect ratio of the image (width:height)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,3)) 
for ax in axes:
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y_axis')


# First Axes 
axes[0].plot(x, y, 'r')
axes[0].set_title('Function')



# Second Axes
axes[1].plot(x, y1, 'g')
axes[1].set_title('First Derivative')

# Third Axes
axes[2].plot(x, y2, 'b')
axes[2].set_title('Second Derivative')

fig.tight_layout() # adjusts subplot(s) params to fit them into the fig area
````

## Adding legends, formatting in LaTeX and fonts

### Legends

A better method is to use the label="label text" keyword argument when plots or other objects are added to the figure, and then using the legend method without arguments to add the legend to the figure:

````python
fig, axes = plt.subplots()
# fig, axes = plt.subplots(figsize=(12,3)) # you can also change the aspect ratio

# plots
axes.plot(x, y, 'r', label="curve 1: y = x^2")
axes.plot(x, x**3, "g", label="curve 2: y = x^1") # adding one more plot

# set title
axes.set_title('title')

# set axes labels
axes.set_xlabel('x')
axes.set_ylabel('y')

# set legends
axes.legend(loc=2) # upper left corner
````

### LaTeX and Fonts

````python
fig, axes = plt.subplots()
# fig, axes = plt.subplots(figsize=(12,3)) # you can also change the aspect ratio

# plots
axes.plot(x, y, 'r', label=r'$f(\alpha) = \alpha^2$')
axes.plot(x, x**3, "g", label=r'$f(\alpha) = \alpha^3$')

# set title
axes.set_title('title')

# set axes labels
axes.set_xlabel(r'$\alpha$', fontsize=15)
axes.set_ylabel(r'$f(x)$', fontsize=15)

# set legends
axes.legend(loc=2, fontsize=15) # upper left corner
````

Alternatively, you can always predefine your font family and font size with the rcParams.update() function:

```python
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'serif'})
```

### Linewidths and Styles

To change the line width, we can use the linewidth or lw keyword argument. The line style can be selected using the linestyle or ls keyword arguments:

````python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="blue", linewidth=0.25)
ax.plot(x, x+2, color="blue", linewidth=0.50)
ax.plot(x, x+3, color="blue", linewidth=1.00)
ax.plot(x, x+4, color="blue", linewidth=2.00)

# possible linestype options ‘-‘, ‘--’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="red", lw=2, linestyle='-')
ax.plot(x, x+6, color="red", lw=2, ls='-.')
ax.plot(x, x+7, color="red", lw=2, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="green", lw=2, ls='--', marker='+')
ax.plot(x, x+10, color="green", lw=2, ls='--', marker='o')
ax.plot(x, x+11, color="green", lw=2, ls='--', marker='s')
ax.plot(x, x+12, color="green", lw=2, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="blue");
````

## Control Over Axis Appearance

### Plot Range

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].set_ylim([0, 60])               # y-axis range
axes[1].set_xlim([2, 5])                # x-axis range
axes[1].set_title("custom axes range");
````

### Logarithmic Scale

````python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
````
